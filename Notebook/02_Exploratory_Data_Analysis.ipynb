## Phase 3: Exploratory Data Analysis

### Objective

The objective of this notebook is to perform exploratory data analysis (EDA) on the employee attrition dataset to discover patterns, trends, and relationships between different features and employee attrition.

This notebook covers:

```python
# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset

df = pd.read_csv("../Dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
# Display first five records

df.head()
```

```python
# Target variable distribution

df["Attrition"].value_counts()
```

```python
# Target variable percentage

df["Attrition"].value_counts(normalize=True)*100
```

### Interpretation

The target variable is imbalanced.

Approximately **84%** of employees stayed with the company, whereas only **16%** left.

Since the dataset is imbalanced, relying only on accuracy may produce misleading results. Therefore, evaluation metrics such as Precision, Recall, F1-score, and ROC-AUC will also be used during model evaluation.

## Target Variable Visualization

Visualizing the distribution of the target variable helps us better understand the class imbalance observed above.

```python
# Set plot style

sns.set_style("whitegrid")

# Create the count plot

plt.figure(figsize=(6,4))

sns.countplot(
    x="Attrition",
    data=df
)

# Add title and labels

plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")

# Display the plot

plt.show()
```

#### Interpretation

The count plot clearly shows that employees who stayed with the company significantly outnumber those who left.

This confirms the imbalance observed earlier through numerical analysis. Such imbalance should be considered during model evaluation because a model predicting only the majority class may still achieve high accuracy while failing to identify employees likely to leave.

## Age Distribution

Understanding the age distribution of employees helps identify whether the workforce is concentrated around a particular age group and whether age could potentially influence employee attrition.

```python
# Create a histogram to visualize employee age distribution

plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

# Add title and labels

plt.title("Distribution of Employee Age")
plt.xlabel("Age")
plt.ylabel("Number of Employees")

# Display the plot

plt.show()
```

#### Interpretation

The age distribution indicates that most employees fall within the middle-age group, with fewer employees at very young or older ages.

This suggests that the organization has a predominantly experienced workforce. Further analysis will determine whether attrition differs across different age groups.

## Age vs Attrition

This section examines whether employee age is associated with attrition by comparing the age distribution of employees who stayed and those who left the company.
