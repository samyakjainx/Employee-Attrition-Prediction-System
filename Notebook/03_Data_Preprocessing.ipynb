### Objective

The objective of this notebook is to prepare the employee attrition dataset for machine learning.

This notebook includes:

- Importing required libraries
- Loading the dataset
- Creating a working copy
- Handling missing values

- Encoding categorical variables

```python
# Import required libraries

import pandas as pd
import numpy as np
import joblib

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

## Load Dataset

Load the Employee Attrition dataset into a pandas DataFrame.

```python
# Load dataset

df = pd.read_csv("../Dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Display first five rows

df.head()
```

## Create a Working Copy

A separate copy of the dataset is created so that preprocessing operations do not modify the original data.

```python
# Create a copy of the dataset

data = df.copy()

# Display shape

data.shape
```

## Check Missing Values

Verify whether the dataset contains any missing values before applying preprocessing techniques.

```python
# Missing values

data.isnull().sum()
```

#### Interpretation

No missing values are present in the dataset.

Therefore, no imputation techniques are required.

## Check Duplicate Records

Verify whether the dataset contains duplicate records before applying preprocessing techniques.

```python
# Check duplicate rows

duplicate_rows = data.duplicated().sum()

print(f"Total Duplicate Records: {duplicate_rows}")
```

#### Interpretation

No duplicate records were found in the dataset.

Therefore, duplicate removal is not required.

## Identify Constant Columns

Columns containing only one unique value do not provide useful information for machine learning models.

Such columns are identified before removal.

```python
# Identify constant columns

constant_columns = [
    column
    for column in data.columns
    if data[column].nunique() == 1
]

print("Constant Columns:")
print(constant_columns)
```

#### Interpretation

Three columns contain only a single unique value throughout the dataset.

Since these features do not contribute any predictive information, they will be removed.

## Remove Unnecessary Features

Remove identifier columns and constant columns that do not contribute to employee attrition prediction.

```python
# Drop unnecessary columns

columns_to_drop = constant_columns + ["EmployeeNumber"]

data.drop(columns=columns_to_drop, inplace=True)

print(f"Dataset Shape after Preprocessing: {data.shape}")
```

```python
data.head()
```

#### Interpretation

Identifier and constant columns have been removed successfully.

The remaining features contain useful information for predictive modeling.

# Handle Categorical Variables

Machine learning algorithms cannot directly process text values.

Therefore, all categorical features are converted into numerical representations using Label Encoding.

This ensures that the dataset is suitable for model training.

### Encode Categorical Features

```python
# Import LabelEncoder

from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataset

data = df.copy()

# Store all encoders

encoders = {}

# Find categorical columns

cat_cols = data.select_dtypes(include="object").columns

# Encode categorical columns

for col in cat_cols:

    le = LabelEncoder()

    data[col] = le.fit_transform(data[col])

    encoders[col] = le

print("Categorical features encoded successfully.")
```

```python
processed_data = data.copy()
```

### Save lable Encoders

All label encoders are saved so the Streamlit application can transform user inputs using exactly the same mappings used during training.

```python
# Save all encoders

joblib.dump(encoders, "../Model/encoders.pkl")

print("Encoders saved successfully.")
```

```python
processed_data.head()
```

#### Interpretation

All categorical variables have been successfully converted into numerical values.

The dataset is now compatible with machine learning algorithms that require numerical input.

## Verify Encoded Dataset

Display the first five records of the encoded dataset to confirm that all categorical variables have been successfully converted into numerical values.

```python
# Display the encoded dataset

processed_data.head()
```

## Verify Data Types

Verify that all features are now represented using numerical data types.

```python
# Display data types

processed_data.info()
```

# Feature Scaling

The numerical features in the dataset have different value ranges.

Feature scaling standardizes these values so that each feature contributes equally during model training, improving the performance of many machine learning algorithms.

```python
processed_data.dtypes
```

```python
# Import StandardScaler

from sklearn.preprocessing import StandardScaler
```

## Separate Features and Target Variable

Separate the independent features (X) from the target variable (y) before applying feature scaling.

```python
# Separate features and target

X = processed_data.drop("Attrition", axis=1)
y = processed_data["Attrition"]

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

## Apply Feature Scaling

Standardize the feature values using StandardScaler.

Only the independent features are scaled, while the target variable remains unchanged.

```python
print(X.select_dtypes(include="object").columns)
```

```python
print(processed_data.head())
```

```python
print(processed_data.dtypes)
```

```python
# Initialize StandardScaler

scaler = StandardScaler()

# Scale the features

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(
    X_scaled,
    columns=X.columns
)
```

###Save Feature Scaler

The trained scaler is saved to ensure that user inputs are standardized in the same way as the training dataset.

```python
# Save scaler

joblib.dump(scaler, "../Model/scaler.pkl")

print("Scaler saved successfully.")
```

### Save Feature Order

The feature order is saved so that the Streamlit application passes the input variables to the trained model in the exact sequence expected.

```python
# Save feature order

joblib.dump(X.columns.tolist(), "../Model/features.pkl")

print("Feature order saved successfully.")
```

```python
# Convert scaled data back into a DataFrame

X_scaled = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

X_scaled.head()
```

#### Interpretation

The independent features have been standardized using StandardScaler.

Each feature now has a mean close to 0 and a standard deviation close to 1, ensuring that variables with larger numerical ranges do not dominate the learning process.

The target variable was intentionally excluded from scaling because it represents the class labels for prediction.

# Save Processed Dataset

The preprocessed dataset is saved as a new CSV file.

This allows the cleaned and transformed data to be reused in later stages of the project without repeating preprocessing.

```python
processed_data = X_scaled.copy()

processed_data["Attrition"] = y
```

```python
# Combine scaled features and target variable

processed_dataset = X_scaled.copy()

processed_dataset["Attrition"] = y.values

# Display first five rows

processed_dataset.head()
```

```python
# Save processed dataset

processed_data.to_csv(
    "../Dataset/Processed_Employee_Attrition.csv",
    index=False
)
print("Processed dataset saved successfully.")
```
