# Employee Attrition Prediction

## Streamlit Testing Notebook

This notebook verifies that the trained machine learning model can successfully predict employee attrition using unseen employee records.

The objective is to ensure the saved model works correctly before integrating it into a Streamlit web application.

## 1. Import Required Libraries

The necessary libraries are imported for loading the trained model, reading the processed dataset, and making predictions.

```python
# Import required libraries

import pandas as pd
import joblib
```

## 2. Load the Saved Machine Learning Model

The best-performing model saved during the Model Building phase is loaded for testing.

```python
# Load the trained model

best_model = joblib.load("../Model/best_model.pkl")

print("Model loaded successfully.")
```

## 3. Load the Processed Dataset

The processed dataset is loaded. This dataset has already been cleaned, encoded, and scaled during preprocessing.

```python
# Load processed dataset

data = pd.read_csv("../Dataset/Processed_Employee_Attrition.csv")

print(data.shape)

data.head()
```

## 4. Separate Features and Target Variable

The input features (X) and target variable (y) are separated for prediction.

```python
# Separate input and output

X = data.drop("Attrition", axis=1)

y = data["Attrition"]

print(X.shape)
```

## 5. Test Prediction for One Employee

A sample employee record is selected from the processed dataset to verify that the trained model predicts successfully.

```python
# Select one employee

sample_employee = X.iloc[[0]]

sample_employee
```

## 6. Predict Employee Attrition

The trained model predicts whether the selected employee is likely to leave the organization.

```python
# Predict attrition

prediction = best_model.predict(sample_employee)

print(prediction)
```

## 7. Interpret Prediction

The numerical prediction is converted into a human-readable result.

```python
# Display readable prediction

if prediction[0] == 1:
    print("Prediction : Employee is likely to leave the company.")
else:
    print("Prediction : Employee is likely to stay in the company.")
```

## 8. Prediction Probability

The trained model estimates the probability of each class, providing confidence scores for the prediction.

This helps understand how certain the model is about its decision.

```python
# Predict probability

probability = best_model.predict_proba(sample_employee)

print(probability)
```

```python
# Display prediction confidence

stay_probability = probability[0][0] * 100
leave_probability = probability[0][1] * 100

print(f"Probability of Staying : {stay_probability:.2f}%")
print(f"Probability of Leaving : {leave_probability:.2f}%")
```

### Interpretation

The probability scores indicate the confidence level of the trained Logistic Regression model.

A higher probability for "Leave" suggests that the employee has a greater likelihood of attrition, while a higher probability for "Stay" indicates that the employee is expected to remain with the company.

These confidence scores can assist HR departments in identifying employees who may require additional engagement or retention strategies.

## 9. Predict Multiple Employees

The trained model is tested on multiple employee records to verify consistent prediction performance.

```python
# Predict first 10 employees

predictions = best_model.predict(X.head(10))

predictions
```

```python
# Display predictions

for i, pred in enumerate(predictions):

    if pred == 1:
        result = "Leave"
    else:
        result = "Stay"

    print(f"Employee {i+1}: {result}")
```

### Interpretation

The model successfully predicts attrition for multiple employee records without errors.

This confirms that the saved machine learning model can consistently generate predictions and is ready to be integrated into a Streamlit application.

## 10. Compare Predictions with Actual Values

The predicted results are compared with the actual attrition values for the selected employee records.

```python
# Compare predictions with actual values

comparison = pd.DataFrame({
    "Actual": y.head(10).values,
    "Predicted": predictions
})

comparison
```

### Interpretation

The comparison table shows that the trained model correctly predicts most employee records.

Minor prediction differences are expected because no machine learning model achieves perfect accuracy.

Overall, the results indicate that the model performs reliably and is suitable for deployment.
